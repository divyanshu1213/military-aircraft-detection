```python
# Since we don't have a real dataset, we simulate one
# This is standard practice for demo/portfolio projects

import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential

# Simulate dataset (64x64 RGB images, 5 classes of aircraft)
print("Creating dataset...")

NUM_CLASSES = 5
IMG_SIZE = 64
NUM_TRAIN = 500
NUM_TEST = 100

# Generate random image data (simulates real aircraft images)
X_train = np.random.random((NUM_TRAIN, IMG_SIZE, IMG_SIZE, 3)).astype("float32")
y_train = np.random.randint(0, NUM_CLASSES, NUM_TRAIN)

X_test = np.random.random((NUM_TEST, IMG_SIZE, IMG_SIZE, 3)).astype("float32")
y_test = np.random.randint(0, NUM_CLASSES, NUM_TEST)

# Normalize pixel values
X_train = X_train / 255.0
X_test = X_test / 255.0

print(f"Training samples: {NUM_TRAIN}")
print(f"Test samples: {NUM_TEST}")
print(f"Image shape: {IMG_SIZE}x{IMG_SIZE}x3")
print(f"Number of classes: {NUM_CLASSES}")
print("Dataset ready!")
```

```python
# Build the CNN Model
print("Building CNN model...")

model = Sequential([
    # First Convolutional Block
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    layers.BatchNormalization(),
    layers.MaxPooling2D(2, 2),

    # Second Convolutional Block
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(2, 2),

    # Third Convolutional Block
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(2, 2),

    # Fully Connected Layers
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(NUM_CLASSES, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
print("\nModel built successfully!")
```

```python
# Train the Model
print("Training model... this will take 1-2 minutes")

history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

# Evaluate on test data
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"\nTest Accuracy: {test_accuracy * 100:.2f}%")
print(f"Test Loss: {test_loss:.4f}")
print("\nTraining complete!")
```

```python
# Plot Training Results
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Accuracy plot
ax1.plot(history.history['accuracy'], label='Train Accuracy', color='blue')
ax1.plot(history.history['val_accuracy'], label='Val Accuracy', color='orange')
ax1.set_title('Model Accuracy')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True)

# Loss plot
ax2.plot(history.history['loss'], label='Train Loss', color='blue')
ax2.plot(history.history['val_loss'], label='Val Loss', color='orange')
ax2.set_title('Model Loss')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True)

plt.suptitle('Military Aircraft Detection - Training Results', fontsize=14)
plt.tight_layout()
plt.savefig('training_results.png', dpi=150, bbox_inches='tight')
plt.show()

print("Graph saved as training_results.png!")
```

```python
# Save the model
model.save('aircraft_detection_model.h5')
print("Model saved!")

# Save model summary to a text file
import io
import sys

stream = io.StringIO()
model.summary(print_fn=lambda x: stream.write(x + '\n'))
summary_string = stream.getvalue()
stream.close()

with open('model_summary.txt', 'w') as f:
    f.write(summary_string)
print("Model summary saved!")

# Show all files we created
import os
files = os.listdir('.')
print("\nFiles created:")
for f in files:
    print(f"  - {f}")
```
